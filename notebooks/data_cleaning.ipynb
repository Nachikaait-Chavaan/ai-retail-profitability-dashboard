```python
import pandas as pd
sales = pd.read_csv('../data/raw/sales_data.csv')
product = pd.read_csv('../data/raw/product_data.csv')
region = pd.read_csv('../data/raw/region_data.csv')
print("Sales:")
display(sales.head())

print("Product:")
display(product.head())

print("Region:")
display(region.head())
# Merge sales with product data on 'Product'
df = pd.merge(sales, product, on='Product', how='left')

# Merge with region data on 'Region'
df = pd.merge(df, region, on='Region', how='left')
df['Revenue'] = df['Price'] * df['Quantity']
df['Cost'] = df['Cost Price'] * df['Quantity']
df['Profit'] = df['Revenue'] - df['Cost']
df['Margin %'] = round((df['Profit'] / df['Revenue']) * 100, 2)
df.head()
df.to_csv('../data/processed/final_data.csv', index=False)
print("✅ Final cleaned dataset saved to data/processed/final_data.csv")
```
